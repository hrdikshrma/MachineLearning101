```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
from sklearn.datasets import load_breast_cancer
pd.options.display.max_columns = None
```

## Step 1 : Loading the breast cancer dataset

```python
cancer_dataset=load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
print(cancer_dataset.DESCR)
```

```python
data = pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names'])
data.head()
```

## Step 2 : Apply Standard Scaler

```python
## Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

## Step 3 : Apply PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=None)
data_pca=pca.fit_transform(scaled_data)
pca.explained_variance_
```

```python
pca = PCA(n_components=10)
data_pca=pca.fit_transform(scaled_data)
print(pca.explained_variance_)
print("Total Variance Explained by the principal components : ", sum(pca.explained_variance_))
```

```python
data_pca.shape
```

```python
reduced_data = pd.DataFrame(data_pca)
```
