```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

## Step 1 : Taking a sample data

```python
# Quadratic Equation used 
X = 6 * np.random.rand(100,1) - 3
y = 0.5 * X**2 + 1.5*X + 2 + np.random.rand(100,1)
plt.scatter(X,y,color='g')
plt.xlabel('X')
plt.ylabel('y')
```

## Step 2 : Dividing data into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)
```

## Step 3 : (Optional) Try Linear Regression to see how much it difference we get

```python
from sklearn.linear_model import LinearRegression
regression_1 = LinearRegression()
regression_1.fit(X_train,y_train)
```

## Step 4 : (Optional) Calculate the performance metrics for Linear Regression

```python
# Prediciton for test data
y_pred = regression_1.predict(X_test)
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
score = r2_score(y_test,y_pred)
print(mae)
print(mse)
print(rmse)
print(score)
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, regression_1.predict(X_train), color = 'red')
plt.xlabel('X')
plt.ylabel('y')
```

## Step 5 : Applying Polynomial Transformation

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

regression_2 = LinearRegression()
regression_2.fit(X_train_poly,y_train)

y_pred = regression_2.predict(X_test_poly)


mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
score = r2_score(y_test,y_pred)
print(mae)
print(mse)
print(rmse)
print(score)
```

```python
plt.scatter(X_train, y_train)
plt.scatter(X_train, regression_2.predict(X_train_poly), color = 'red')
plt.xlabel('X')
plt.ylabel('y')
```
